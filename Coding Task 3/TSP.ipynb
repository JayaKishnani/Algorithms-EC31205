```python
import math as mt 

def CopyToFinal(curr_path):
    final_path[:Num_Node + 1] = curr_path[:]
    final_path[Num_Node] = curr_path[0]

def FirstMin(adj_matrix, i):
    min = float('inf')
    for k in range(Num_Node):
        if adj_matrix[i][k] < min and i != k:
            min = adj_matrix[i][k]
  
    return min

def SecondMin(adj_matrix, i):
    first, second = float('inf'), float('inf')
    for j in range(Num_Node):
        if i == j:
            continue
        if adj_matrix[i][j] <= first:
            second = first
            first = adj_matrix[i][j]
  
        elif(adj_matrix[i][j] <= second and 
             adj_matrix[i][j] != first):
            second = adj_matrix[i][j]
    return second

def TSPRec(adj_matrix, curr_bound, curr_weight, level, current_path, visit):
    global final_min_cost

    if level == Num_Node:

        if adj_matrix[current_path[level - 1]][current_path[0]] != 0:

            curr_res = curr_weight + adj_matrix[current_path[level - 1]]\
                                        [current_path[0]]
            if curr_res < final_min_cost:
                CopyToFinal(current_path)
                final_min_cost = curr_res
        return

    i = 0
    while(i<Num_Node):

        if (adj_matrix[current_path[level-1]][i] != 0 and
                            visit[i] == False):
            temp_elemnet = curr_bound
            curr_weight += adj_matrix[current_path[level - 1]][i]

            if level == 1:
                curr_bound -= ((FirstMin(adj_matrix, current_path[level - 1]) + 
                                FirstMin(adj_matrix, i)) / 2)
            else:
                curr_bound -= ((SecondMin(adj_matrix, current_path[level - 1]) +
                                 FirstMin(adj_matrix, i)) / 2)
  

            if curr_bound + curr_weight < final_min_cost:
                current_path[level] = i
                visit[i] = True

                TSPRec(adj_matrix, curr_bound, curr_weight, 
                       level + 1, current_path, visit)

            curr_weight -= adj_matrix[current_path[level - 1]][i]
            curr_bound = temp_elemnet

            visit = [False] * len(visit)
            for j in range(level):
                if current_path[j] != -1:
                    visit[current_path[j]] = True
        i+=1
  
def TSP_main_func(adj_matrix):

    curr_bound = 0
    curr_path = [-1] * (Num_Node + 1)
    visit = [False] * Num_Node

    for i in range(Num_Node):
        curr_bound += (FirstMin(adj_matrix, i) + 
                       SecondMin(adj_matrix, i))

    curr_bound = mt.ceil(curr_bound / 2)

    visit[0] = True
    curr_path[0] = 0

    TSPRec(adj_matrix, curr_bound, 0, 1, curr_path, visit)

adj_matrix = [[0, 1 , 4 , 0, 0 , 0 ,0 ],
       [1, 0 , 0 , 3, 6 , 0 ,0 ],
       [4, 0 , 0 , 2, 0 , 5 ,0 ],
       [0, 3 , 2 , 0, 2 , 4 ,0 ],
       [0, 6 , 0 , 2, 0 , 2 ,7 ],
       [0, 0 , 5 , 4, 2 , 0 ,6 ],
       [0, 0 , 0 , 0, 7 , 6 ,0 ]
      ]
Num_Node = 7

final_path = [None] * (Num_Node + 1)

visit = [False] * Num_Node

final_min_cost = float('inf')
  
TSP_main_func(adj_matrix)
  
print("Minimum cost :", final_min_cost)
print("Path Taken : ", end = ' ')
i = 0
while(i<Num_Node):
    print(final_path[i]+1, end = ' ')
    i+=1
```
